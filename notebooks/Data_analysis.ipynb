```python
import os
import random
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Path to the directory containing tumor files
tumor_files_path = r"C:\Users\Devshree Deshmukh\OneDrive\Desktop\MSc Project\BraTS_ML_Project\data\Tumour Data"

# Function to determine the axial view dimension
def get_axial_view_dimension(affine):
    # The affine matrix maps voxel indices to real-world coordinates
    # The axis with the largest absolute value in the third row corresponds to the axial view
    return np.argmax(np.abs(affine[:3, 2]))

# Get a list of all NIfTI files in the directory
nifti_files = [f for f in os.listdir(tumor_files_path) if f.endswith(".nii") or f.endswith(".nii.gz")]

# Select 3 random files
random_files = random.sample(nifti_files, 3)

# Visualize 5 slices before and 40 slices after the middle slice
for file_name in random_files:
    file_path = os.path.join(tumor_files_path, file_name)
    
    # Load the NIfTI file
    img = nib.load(file_path)
    data = img.get_fdata()
    affine = img.affine
    
    # Determine the axial view dimension
    axial_dim = get_axial_view_dimension(affine)
    
    # Get the middle slice in the axial view
    middle_slice_index = data.shape[axial_dim] // 2
    
    # Determine the range of slices to visualize
    start_slice = max(0, middle_slice_index - 5)
    end_slice = min(data.shape[axial_dim], middle_slice_index + 41)  # +41 to include 40 slices after the middle slice
    
    # Plot the slices
    num_slices = end_slice - start_slice
    fig, axes = plt.subplots(2, (num_slices + 1) // 2, figsize=(25, 10))  # Adjust layout for better visibility
    fig.suptitle(f"File: {file_name} - Slices around the middle slice", fontsize=16)
    
    axes = axes.flatten()  # Flatten axes for easy indexing
    for i, slice_index in enumerate(range(start_slice, end_slice)):
        if axial_dim == 0:
            slice_data = data[slice_index, :, :]
        elif axial_dim == 1:
            slice_data = data[:, slice_index, :]
        else:
            slice_data = data[:, :, slice_index]
        
        axes[i].imshow(slice_data.T, cmap="gray", origin="lower")
        axes[i].set_title(f"Slice {slice_index}")
        axes[i].axis("off")
    
    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis("off")
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to fit the title
    plt.show()
```
